```python
import numpy as np

def init_network():
    network ={}
    network['bias'] = 1
    network['weight1'] = np.random.random() #w1
    network['weight2'] = np.random.random() #w2
    return network

def train(network, x1, x2):
    y = np.array([0,0,0,1])
    for i in range(100):
        for x_,y_,t_ in zip(x1,x2,y):
            w = np.array([network['weight1'], network['weight2']])
            arr = np.array([x_, y_])
            sig = np.sum(arr*w) +  network['bias']
            up = 0.5 * MSELoss(sigmoid(sig) , t_)
            network['weight1'] = network['weight1']  -  (up * x_)
            network['weight2'] = network['weight2']  - (up * y_)
            network['bias'] = network['bias'] - up
            print("Weight 1 : {}, Weight 2 : {} , Bias : {}".format(network['weight1'] ,network['weight2'] , network['bias']))            
    return network

def test(network, x1, x2):
    label = []
    for x, y in zip(x1,x2):
        arr = np.array([x, y])
        w = np.array([network['weight1'], network['weight2']])
        sig = np.sum(arr*w) + network['bias']
        if sigmoid(sig) >= 0.5:
            label.append(1)
        else:
            label.append(0)
    return label


def MSELoss(predict, target):
    
    return (1/2) * np.sum((predict-target)**2)

def sigmoid(x):
    
    return 1/(1 + np.exp(-x))

if __name__ == '__main__':
    Data = init_network();
    Data = train(Data, np.array([0,1,0,1]), np.array([0,0,1,1]))
    Data = test(Data, np.array([0,1,0,1]), np.array([0,0,1,1]))
    print((np.array(Data)))
```
