```python
import pandas as pd
import numpy as np
from sklearn import cluster
import matplotlib.pyplot as plt
from sklearn.metrics.cluster import adjusted_rand_score

colors = np.array(['red', 'green', 'blue', 'pink', 'aqua','orange'])
df = pd.read_csv('lab8_test.csv', index_col = -1)
for i in pd.DataFrame(df):
    df[i],class_ = pd.factorize(df[i], sort=True)
features = pd.DataFrame(df)



k = 3
cluster = cluster.KMeans(n_clusters= k)
cluster = cluster.fit(features)
pre = cluster.predict(features)

plt.title("k = 3 ARI = {} ".format(adjusted_rand_score(pre, features.index)))
plt.scatter(features['sepal length'], features['petal length'], color = colors[pre])


```
